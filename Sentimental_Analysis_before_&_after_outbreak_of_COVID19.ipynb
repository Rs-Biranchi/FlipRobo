```python
def get_tweets():
    import GetOldTweets3 as got
    tweetCriteria=got.manager.TweetCriteria().setQuerySearch('CoronaVirus') \
        .setSince("2019-11-28")\
        .setUntil('2019-12-28')\
        .setLang('english')\
        .setMaxTweets(500)
    
    tweets=got.manager.TweetManager.getTweets(tweetCriteria)
    text_tweets=[[tweet.text] for tweet in tweets]
    return text_tweets
```

```python
text=''
text_tweets=get_tweets()

length=len(text_tweets)

for i in range(0,length):
    text=text_tweets[i][0]+" "+text
    
print(text)
```

```python
#saving the document as list in A
A=[text]
lower_case=text.lower()
cleaned_text=lower_case.translate(str.maketrans('','',string.punctuation))

tokenized_words=cleaned_text.split()
tokenized_words
```

```python
from nltk.corpus import stopwords
stopword=stopwords.words()
stopword
```

```python
Final_words=[word for word in tokenized_words if word not in stopword]
Final_words
```

```python
Emotion_list=[]

with open('emotions.txt','r') as file:
    for line in file:
        clear_line=line.replace('\n','').replace(',','').replace("'",'').strip()
        word,emotion=clear_line.split(':')
        if word in Final_words:
            Emotion_list.append(emotion)
            
w=Counter(Emotion_list) 
print(w)
```

```python
#Sentiment before outbreak of coronavirus
fig, ax1=plt.subplots()
ax1.bar(w.keys(),w.values())
fig.autofmt_xdate()
plt.savefig('graph.png')
plt.show()
```

```python
def Get_tweets():
    import GetOldTweets3 as got
    tweetCriteria=got.manager.TweetCriteria().setQuerySearch('CoronaVirus')\
    .setSince("2020-03-15")\
    .setUntil("2020-04-14")\
    .setMaxTweets(500)
    
    tweets=got.manager.TweetManager.getTweets(tweetCriteria)
    text_tweets=[[tweet.text] for tweet in tweets]
    return text_tweets
```

```python
#reading text files
text=""
Text_tweets=Get_tweets()
length=len(Text_tweets)

for i in range(0,length):
    text=Text_tweets[i][0]+" "+text
    
print(text)    
```

```python
#saving the document as list in B
B=[text]
#converting to lower_case
lower_case=text.lower()

#removing punctuations
cleaned_text=lower_case.translate(str.maketrans('',"",string.punctuation))

#splitting text into words
tokenized_words=cleaned_text.split()
print(tokenized_words)
```

```python
Final_words=[word for word in tokenized_words if word not in stopword]
Final_words
```

```python
#Get Impotions Text
Emotion_list2=[]
with open('emotions.txt','r') as file:
    for line in file:
        clear_line=line.replace('\n','').replace(',','').replace("'",'').strip()
        word,emotion=clear_line.split(':')
        if word in Final_words:
            Emotion_list2.append(emotion)
            
W=Counter(Emotion_list2)
print(W)
```

```python
#sentiment after Outbreak of CoronaVirus
figg,ax1=plt.subplots()
ax1.bar(W.keys(),W.values())
figg.autofmt_xdate()
plt.savefig('graph.png')
plt.show()
```

```python
#Check How many words of B are present in BOW of A
from sklearn.feature_extraction.text import CountVectorizer
vect1=CountVectorizer()
vect1.fit_transform(A) 
print("Bag of Words of A",vect1.get_feature_names())
```

```python
vect2=CountVectorizer()
vect2.fit(A)
```

```python
#Transform B
new_vect=vect2.transform(B)

print("Text present at: ",new_vect.toarray())
```

```python
#We can see the indexes where the words of B are present in Bag of words of A. 
import pandas as pd
df=pd.DataFrame(new_vect.toarray())
df.to_csv('Text_present_at.csv')
```
