```
import pandas as pd
import numpy as np
import tensorflow as tf
```

```
data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/titanic.csv')
data
```

```
data.info()
```

```
data.drop(['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)
data.head()
```

```
data.isnull().sum()
```

```
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy='mean')
data.Age=imp.fit_transform(data.Age.values.reshape(-1,1))
data.isnull().sum()
```

```
imp=SimpleImputer(strategy='most_frequent')
data.Embarked=imp.fit_transform(data.Embarked.values.reshape(-1,1))
data.isnull().sum()
```

```
data.nunique()
```

```
from sklearn.preprocessing import LabelEncoder
```

```
data.Sex=pd.get_dummies(data['Sex'],drop_first=True)
data.head()
```

```
le=LabelEncoder()
data.Embarked=le.fit_transform(data.Embarked.values.reshape(-1,1))
data.head()
```

```
x=data.iloc[:,1:10]
x.head()
```

```
y=data.iloc[:,0].values.reshape(-1,1)
```

```
from sklearn.model_selection import train_test_split
train_x,train_y,test_x,test_y=train_test_split(x,y,test_size=0.25,stratify=y)
```
