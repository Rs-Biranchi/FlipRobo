```python
#data preprocessing
Text=re.sub(r'\[[0-9]*\]',' ',text)
Text_1=re.sub(r'\s+',' ',Text)

Text_2=Text_1.lower()

Final_text=re.sub(r'\d',' ',Text_2)
print(Final_text)
```

```python
#removing punctuations 
import string
final_text=Final_text.translate(str.maketrans("","",string.punctuation))
final_text
```

```python
#sentence tokenization
sent_tok=nltk.sent_tokenize(final_text)
#word tokenization of sent_tok
word_tok=[nltk.word_tokenize(sent) for sent in sent_tok]

from nltk.corpus import stopwords

for i in range(len(word_tok)):
    word_tok[i]=[word for word in word_tok[i]
                 if word not in stopwords.words('english')]

print(word_tok)
```

```python
#Training the Word2Vec   ['dynasty','chinese','monks','troops','visited','export','vision','hindichini','well','built','road','imperialism','sovereignty','maps','nehru','communist','reassert','british','friendship','policy']
model=Word2Vec(word_tok,min_count=1)
words=model.wv.vocab
```

```python
vec=model.wv['india']
vec
```

```python
#Similar
similar=model.wv.most_similar('india',topn=5)
similar
```

```python
#Check similarity of words
list=['dynasty','chinese','monks','troops','visited','export','vision','hindichini','well','built','road','imperialism','sovereignty','maps','nehru','communist','reassert','british','friendship','policy']

for i in list:
    print("Similar words of ",i," ",model.wv.most_similar(i,topn=5))
    print("\n")
```
